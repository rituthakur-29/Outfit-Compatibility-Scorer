```python
mens_df = pd.read_csv(
    '../data/processed/mens_fashion.csv'
)
```

Defining Categories

```python
tops = [
    'Tshirts',
    'Shirts',
    'Kurtas',
    'Tops',
    'Sweatshirts',
    'Sweaters',
    'Jackets'
]

bottoms = [
    'Jeans',
    'Trousers',
    'Shorts',
    'Track Pants'
]

footwear = [
    'Casual Shoes',
    'Sports Shoes',
    'Formal Shoes'
]
```

```python
tops_df = mens_df[
    mens_df['articleType'].isin(tops)
]

bottoms_df = mens_df[
    mens_df['articleType'].isin(bottoms)
]

footwear_df = mens_df[
    mens_df['articleType'].isin(footwear)
]
```

Sample Data

```python
tops_sample = tops_df.sample(100, random_state=42)
bottoms_sample = bottoms_df.sample(100, random_state=42)
footwear_sample = footwear_df.sample(100, random_state=42)
```

```python
import random
```

Outfit Combinations

```python
outfits = []

for _ in range(3000):
    
    top = tops_sample.sample(1).iloc[0]
    bottom = bottoms_sample.sample(1).iloc[0]
    shoe = footwear_sample.sample(1).iloc[0]

    outfits.append({
        'top_type': top['articleType'],
        'top_color': top['baseColour'],
        'top_usage': top['usage'],
        'top_season': top['season'],

        'bottom_type': bottom['articleType'],
        'bottom_color': bottom['baseColour'],
        'bottom_usage': bottom['usage'],
        'bottom_season': bottom['season'],

        'shoe_type': shoe['articleType'],
        'shoe_color': shoe['baseColour'],
        'shoe_usage': shoe['usage'],
        'shoe_season': shoe['season']
    })

outfits_df = pd.DataFrame(outfits)
```
