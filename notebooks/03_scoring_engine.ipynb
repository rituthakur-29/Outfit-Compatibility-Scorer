Load Data

```python
import pandas as pd

outfits_df = pd.read_csv(
    '../data/processed/outfit_combinations.csv'
)
```

```python
import random
```

Implementation of Score

```python
good_color_pairs = {
    ('White', 'Black'),
    ('White', 'Navy Blue'),
    ('Blue', 'Beige'),
    ('Grey', 'Black'),
    ('Brown', 'Beige')
}
```

```python
neutral_colors = [
    'Black',
    'White',
    'Grey',
    'Navy Blue',
    'Brown',
    'Beige'
]

def calculate_score(row):

    score = 0

    # Usage Score
    usages = [
        row['top_usage'],
        row['bottom_usage'],
        row['shoe_usage']
    ]

    if len(set(usages)) == 1:
        score += 40
    elif len(set(usages)) == 2:
        score += 20
    else:
        score += 5

    # Season Score
    seasons = [
        row['top_season'],
        row['bottom_season'],
        row['shoe_season']
    ]

    if len(set(seasons)) == 1:
        score += 20
    elif len(set(seasons)) == 2:
        score += 10

    # Color Score
    colors = [
        row['top_color'],
        row['bottom_color'],
        row['shoe_color']
    ]

    neutral_count = sum(
        color in neutral_colors
        for color in colors
    )

    if neutral_count == 3:
        score += 30
    elif neutral_count == 2:
        score += 20
    else:
        score += 10

    # Bonus for good color combinations

    if (
        row['top_color'],
        row['bottom_color']
    ) in good_color_pairs:

        score += 15

    # Category Logic
    if (
        row['top_type'] == 'Shirts'
        and row['bottom_type'] == 'Trousers'
        and row['shoe_type'] == 'Formal Shoes'
    ):
        score += 10

    elif (
        row['top_type'] == 'Tshirts'
        and row['bottom_type'] == 'Jeans'
        and row['shoe_type'] == 'Casual Shoes'
    ):
        score += 10

    elif (
        row['top_type'] == 'Sweatshirts'
        and row['bottom_type'] == 'Jeans'
        and row['shoe_type'] == 'Casual Shoes'
    ):
        score += 10

    elif (
        row['top_type'] == 'Kurtas'
        and row['bottom_type'] == 'Trousers'
        and row['shoe_type'] == 'Formal Shoes'
    ):
        score += 10

    elif (
        row['top_type'] == 'Tshirts'
        and row['bottom_type'] == 'Track Pants'
        and row['shoe_type'] == 'Sports Shoes'
    ):
        score += 10


   # Randomness
    score += random.randint(-3, 3)

    score = max(0, min(100, score))

    return score
```

```python
outfits_df['compatibility_score'] = outfits_df.apply(
    calculate_score,
    axis=1
)
```

```python
outfits_df[
    ['top_type',
     'bottom_type',
     'shoe_type',
     'compatibility_score']
].head(10)
```

```python
def get_label(score):

    if score >= 80:
        return 'Highly Compatible'

    elif score >= 60:
        return 'Moderately Compatible'

    else:
        return 'Poor Match'
```

```python
outfits_df['compatibility_label'] = (
    outfits_df['compatibility_score']
    .apply(get_label)
)
```
