## Tops, Bottoms and Footwears


```python
tops = [
    'Tshirts',
    'Shirts',
    'Kurtas',
    'Tops',
    'Sweatshirts',
    'Sweaters',
    'Jackets'
]

bottoms = [
    'Jeans',
    'Trousers',
    'Shorts',
    'Track Pants'
]

footwear = [
    'Casual Shoes',
    'Sports Shoes',
    'Formal Shoes'
]
```

```python
# Keep only outfit-related items
relevant_items = tops + bottoms + footwear

fashion_df = df[df['articleType'].isin(relevant_items)].copy()
```

```python
fashion_df.shape

fashion_df['articleType'].value_counts()

fashion_df['gender'].value_counts()

fashion_df['usage'].value_counts()
```

```python
fashion_df.to_csv(
    '../data/processed/fashion_products.csv',
    index=False
)
```

```python
fashion_df = df[df['articleType'].isin(relevant_items)]

fashion_df['articleType'].value_counts()
```

```python
mens_df = fashion_df[
    (fashion_df['gender'] == 'Men') &
    (fashion_df['articleType'].isin(
        tops + bottoms + footwear
    ))
].copy()
```
